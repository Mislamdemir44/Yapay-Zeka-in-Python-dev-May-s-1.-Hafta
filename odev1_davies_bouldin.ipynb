```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
```

```python
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Görselleştirelim
plt.scatter(X[:, 0], X[:, 1])
plt.title("Veri Kümesi")
plt.xlabel("Özellik 1")
plt.ylabel("Özellik 2")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Davies-Bouldin skoru
db_score = davies_bouldin_score(X, labels)
print(f"Davies-Bouldin Skoru: {db_score:.3f}")
```

### Davies-Bouldin Skoru Nedir?
- DB skoru, kümelerin birbirine ne kadar benzediğini ölçen bir metriktir.
- Skor **ne kadar düşükse**, kümeler **o kadar ayrık ve yoğun** demektir.
- 0'a yakın değerler iyi kümelemeyi gösterir.
